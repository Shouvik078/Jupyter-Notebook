```python
# Conditional use in Pandas using function

# less than 100    = LOW
# 100 to 500       = MEDIUM
# Greater than 500 = HIGH 

def check(price):
    if price<100:
        return 'LOW'
    elif price<500:
        return 'MEDIUM'
    else:
        return 'HIGH' 
    
```

```python
#making extra column 
df['Price Status']=df['Price_Per_Unit'].apply(check)  
```
